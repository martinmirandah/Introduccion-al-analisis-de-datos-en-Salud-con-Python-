## 1. Sintaxis básica y operaciones

### 1.1 Imprimir resultados (`print`)
La función `print()` permite mostrar resultados en pantalla. Es la herramienta más básica para inspeccionar el resultado de nuestro código.


```python
# Ejemplo básico de print()
print("Hola, bienvenidos a la Unidad 2")
print("La suma de 2 + 3 es:", 2 + 3)
```

### 1.2 Comentarios
Los comentarios (`#`) no se ejecutan; sirven para documentar el código y explicar qué hace cada parte.


```python
# Esto es un comentario, Python lo ignora al ejecutar
x = 10  # Asignamos el valor 10 a la variable x
print(x)  # Mostramos el valor de x
```

### 1.3 Operaciones aritméticas
Python soporta las operaciones aritméticas básicas: suma (`+`), resta (`-`), multiplicación (`*`), división (`/`), división entera (`//`), módulo (`%`) y potencia (`**`).


```python
a = 17
b = 5

print("Suma:", a + b)
print("Resta:", a - b)
print("Multiplicación:", a * b)
print("División:", a / b)
print("División entera:", a // b)
print("Módulo (resto):", a % b)
print("Potencia:", a ** 2)
```

### 1.4 Operadores de asignación
Los operadores de asignación compuestos (`+=`, `-=`, `*=`, `/=`, `**=`) permiten actualizar el valor de una variable de forma más compacta.


```python
total = 0
total += 50   # equivalente a total = total + 50
print("Después de +=:", total)

total -= 10   # equivalente a total = total - 10
print("Después de -=:", total)

total *= 2    # equivalente a total = total * 2
print("Después de *=:", total)

total /= 4    # equivalente a total = total / 4
print("Después de /=:", total)
```

### 1.5 Operadores de comparación
Los operadores de comparación (`==`, `!=`, `>`, `<`, `>=`, `<=`) evalúan una condición y devuelven un valor booleano (`True` o `False`).


```python
print(5 == 5)   # Igual a
print(5 != 3)   # Distinto de
print(7 > 4)    # Mayor que
print(7 < 4)    # Menor que
print(7 >= 7)   # Mayor o igual que
print(3 <= 2)   # Menor o igual que
```

## 2. Variables y tipos de datos

### 2.1 Variables y tipos básicos
Una variable es un nombre que referencia un valor almacenado en memoria. Python es de tipado dinámico: no es necesario declarar el tipo de antemano.


```python
nombre = "Ana"        # str (texto)
edad = 28             # int (entero)
altura = 1.65         # float (decimal)
es_estudiante = True  # bool (booleano)

print(nombre, type(nombre))
print(edad, type(edad))
print(altura, type(altura))
print(es_estudiante, type(es_estudiante))
```

### 2.2 Manipulación de strings (cadenas de texto)
Las cadenas admiten concatenación, repetición, indexación y *slicing* (extracción de subcadenas).


```python
mensaje = "Análisis de datos"

print(mensaje.upper())          # Mayúsculas
print(mensaje.lower())          # Minúsculas
print(mensaje[0])               # Primer carácter
print(mensaje[-1])              # Último carácter
print(mensaje[0:8])             # Slicing: primeros 8 caracteres
print(mensaje + " - Clase 2")   # Concatenación
print(mensaje.replace("datos", "información"))  # Reemplazo
```

### 2.3 Expresiones booleanas y operadores lógicos
Los operadores lógicos `and`, `or` y `not` permiten combinar condiciones.


```python
edad = 20
tiene_carnet = True

print(edad >= 18 and tiene_carnet)   # Ambas condiciones deben ser verdaderas
print(edad < 18 or tiene_carnet)     # Al menos una debe ser verdadera
print(not tiene_carnet)              # Negación
```

### 2.4 Conversión de tipos
Es posible convertir explícitamente entre tipos usando `int()`, `float()`, `str()`, `bool()`.


```python
valor_texto = "42"
valor_entero = int(valor_texto)
valor_decimal = float(valor_entero)

print(valor_entero, type(valor_entero))
print(valor_decimal, type(valor_decimal))
print(str(valor_decimal), type(str(valor_decimal)))
```

### 2.5 Tipos de datos complejos: listas, tuplas, conjuntos y diccionarios

**Listas**: colecciones ordenadas y mutables.


```python
datos = [23, 45, 12, 67, 34]

datos.append(99)          # Agregar un elemento
datos.sort()               # Ordenar
print("Lista ordenada:", datos)
print("Primer elemento:", datos[0])
print("Sublista (slicing):", datos[1:4])
```

**Tuplas**: colecciones ordenadas e inmutables, útiles para datos fijos.

```python
paciente = ("Juan Pérez", 45, "Hipertensión")
print(paciente)
print("Diagnóstico:", paciente[2])
```

**Conjuntos (sets)**: colecciones no ordenadas sin elementos duplicados, útiles para operaciones de conjuntos.

```python
grupo_a = {"glucosa", "colesterol", "presión"}
grupo_b = {"colesterol", "triglicéridos", "presión"}

print("Unión:", grupo_a | grupo_b)
print("Intersección:", grupo_a & grupo_b)
print("Diferencia:", grupo_a - grupo_b)
```

**Diccionarios**: pares clave-valor, ideales para representar registros estructurados.

```python
registro = {
    "nombre": "María",
    "edad": 34,
    "diagnóstico": "Diabetes tipo 2"
}

print(registro["nombre"])
registro["tratamiento"] = "Metformina"  # Agregar nueva clave
print(registro)

for clave, valor in registro.items():
    print(f"{clave}: {valor}")
```

## 3. Estructuras de control: bucles y condicionales

### 3.1 Condicionales (`if`, `elif`, `else`)
Permiten ejecutar distintos bloques de código según se cumpla o no una condición.


```python
presion = 135

if presion < 120:
    print("Presión normal")
elif presion < 140:
    print("Presión elevada")
else:
    print("Hipertensión")
```

### 3.2 Bucle `for`
Permite iterar sobre los elementos de una secuencia (lista, tupla, string, etc.).


```python
valores = [98, 102, 87, 110, 95]

for valor in valores:
    if valor > 100:
        print(valor, "-> fuera de rango")
    else:
        print(valor, "-> dentro de rango")
```

### 3.3 Bucle `while`
Repite un bloque de código mientras se cumpla una condición.


```python
contador = 0

while contador < 5:
    print("Iteración número:", contador)
    contador += 1
```

### 3.4 Comprehensions (listas por comprensión)
Una forma compacta de crear listas a partir de otra secuencia.


```python
valores = [98, 102, 87, 110, 95]

# Forma tradicional con for
fuera_rango = []
for v in valores:
    if v > 100:
        fuera_rango.append(v)
print("Tradicional:", fuera_rango)

# Equivalente con comprensión de listas
fuera_rango_comp = [v for v in valores if v > 100]
print("Comprensión:", fuera_rango_comp)
```

### 3.5 Manejo de errores con `try` / `except`
Permite continuar la ejecución del programa aunque ocurra un error.


```python
datos = ["10", "20", "abc", "30"]

for d in datos:
    try:
        print("Valor convertido:", int(d))
    except ValueError:
        print(f"No se pudo convertir '{d}' a número")
```
